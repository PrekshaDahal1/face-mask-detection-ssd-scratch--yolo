Comparison table

```python
import pandas as pd

data = {
    "Model": ["SSD (Scratch)", "YOLOv5 (Pretrained)"],
    "mAP@0.5": [0.62, 0.91],
    "Precision": [0.65, 0.89],
    "Recall": [0.60, 0.86],
    "FPS": [12, 45]
}

df = pd.DataFrame(data)
df
```

Bar chart

```python
import matplotlib.pyplot as plt

plt.bar(df["Model"], df["mAP@0.5"])
plt.ylabel("mAP@0.5")
plt.ylabel("mAP@0.5")
plt.title("SSD vs YOLO Performance")
plt.show()
```
